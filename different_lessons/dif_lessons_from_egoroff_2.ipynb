### Рекурсия

```python
def recurse(x):
    if x < 4: # Добавляем выход из рекурсии
        print(x)
        recurse(x+1)
        print(x)

recurse(1)
```

```python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n-1)
    
factorial(4)
```

```python
def fibonacci(n):
    if n <= 2:
        return 1
    return fibonacci(n-1) + fibonacci(n-2)

fibonacci(5)
```

```python
# Полиндром - строка которая читается одинаково слева направа и справа налево
def polindrom(str):
    if len(str) == 1 or len(str) == 0:
        return True
    if str[0] != str[-1]:
        return False
    else:
        return polindrom(str[1:-1])

print(polindrom('шалаш'))
    
```

```python
# Мое решение

def task_from_informatics(str):
    if len(str) <= 2:
        print('(', str, ')', end='', sep='')
    else:
        print('(', str[0], end='', sep='')
        task_from_informatics(str[1:-1])
        print(str[-1], ')', end='', sep='')


task_from_informatics('hello world')
```

```python
def video_version(str):
    if len(str) <= 2:
        return str
    else:
        return str[0] + '(' + video_version(str[1:-1]) + ')' + str[-1]
    
print(video_version('hello world'))
```

```python
def task_new(str):
    if len(str) == 0:
        return str
    else:
        new_str = str[0]
        new_str += task_new(str[1:])
        new_str.replace('(', ')')
        new_str += str[0]
        # return str[0] + task_new(str) + str.replace('(', ')') + str[0]
        return new_str
    
print(task_new('he((((llo world!'))
```

```python
def power(x, n): # Возведения в степень мое решение
    if n == 0:
        return 1
    return x * power(x, n-1)

power(3, 2)


def power_v2(x, n):  # Решение по видео (не смог продумать)
    if n == 0:
        return 1
    if n < 0:
        return 1/power_v2(x, -n)
    if n%2 == 0:
        return power_v2(x, n//2)*power_v2(x, n//2)
    else:
        return power_v2(x, n-1)*x
    
power_v2(5,2)
```

```python
def check_files(path, level=1):
    print('Level = ',level, path, 'content: ', os.listdir(path))
    for i in os.listdir(path):
        if os.path.isdir(path + '\\' + i):
            print('Спускаемся', path + '\\' + i)
            check_files(path + '\\' + i, level+1)
            print('Возвращаемся')
            
check_files('C:\\Bacek\\DS')
```
