## Опережающее тестирование (методика TDD)

### Основная мысль

Программист знающий программу - не может написать сам нормальные тесты...   
При соблюдении некоторых условий : реально и правильно предварительно писать тесты)))   
Итого выходит что в начале пишем тест, а потом уже пишем функцию и только потом заниматься разработкой...   
Такая идея и называется опережающее тестирование или `test first`  
Т.е. в начале выкладываем себе грабли, а потом пишем функии / методы

### Разработка через тестирование

- `Test-driven Development` - итеративная методика разработки программ, в которой опережающее тестирование `test-first` ставится во главу угла
- Тестированию не просто отдается большой приоритет, но оно управляет процессом дизайна программного продукта

### Вопросы дизайна возникаюшие при разработке тестов 


- Каковы обязанности тестируемой системы?
- Что и когда она должна делать?
- Какой `API` для выполнения задуманной функциональности
- Что конкретно нужно тестируемой системе для выполнения обязательств
- Что мы имеем на выходе?
- Какие есть побочные эффекты работы?
- Как узнать, что система работает правильно
- Достаточно ли хорошо определена эта правильность

Сначало программист становится `заказчиком-критиком` Требования превращаются в тест -> модульный тест

Преимущество `TDD`:
- Эффективное совмещение ролей(собственного кода)
- Рефакторинг без риска испортить собственный кода
- Повышает уверенность в качестве программного кода
Опытные программисты говорят что `TDD` позволяет не нервничать

Прежде чем писать тест, продумываем какие тесты должна проходить функция Фибоначи
- `Simple` Простые:
  - 0 -> 0
  - 1 -> 1
  - 2 -> 1
  - 5 -> 5
- `Stress` Стрессовые (на которых будет долго отрабатывать)
  - 50
  - 100 (возникает вопрос, а как проверить вопрос)
  - 100000 -> `Exception` (value error) должно возникать исключение
- `Negative` На фигню
  - -1 -> 0
  - -100 -> 0 
- `Wrong_types` Не адекватные значения
  - 'lalala' -> отказ от выполнения (а какая кстати ошибка) `ТуpeError`
  - 3.14 -> отказ от выпонения `ТуpeError`
